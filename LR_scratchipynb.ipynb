```python
#lets make a linear regression model that predicts the amount of glucose
#from scratch
X = df.drop('Glucose', axis=1)
y = df['Glucose']

# lets add bias ... now we have n+1 features
X = np.c_[np.ones(X.shape[0]), X]

#normalize the features..puts all features in a similar range
#standardization or z score normalization
X = (X - np.mean(X, axis=0)) / (np.std(X, axis=0)+1e-8) #added a small value to avoid divide by 0
                                                        #if a col has same values for all rows std = 0

theta = np.zeros(X.shape[1])
learning_rate = 0.01
epoch = 100

for _ in range(epoch):
  pred = X.dot(theta)
  error = pred - y
  gradient = X.T.dot(error) / len(y) #here we take the transpose so that multplication is valid
  theta -= learning_rate * gradient
```

```python
theta.shape
print(theta)
```

```python
def predict(X_new,theta):
  #always add the bias
  X_new = np.c_[np.ones[X.shape[0],1],X_new]
  return X_new.dot(theta)
```

```python
#evaluating error
import sklearn
import math
from sklearn.metrics import mean_squared_error
y_pred = X.dot(theta)
rmse = math.sqrt(mean_squared_error(y,y_pred))
print(rmse)
```

```python
#since the mean of glucose is 120 and rmse is 123
rel_error = rmse/df['Glucose'].mean()
print(rel_error * 100)
```
